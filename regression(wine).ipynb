```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('winequality_white.csv')
```

```python
print(df)
```

```python
#check the correlation between dependent variable and independent variable

import seaborn as sns
sns.heatmap(df.corr(),annot=True)

#light color indicates the correlation
```

```python
#seperate independent variable and dependent variable


y=df['quality']
df.drop('quality',axis=1,inplace=True)
x=df
print(x)
print(y)
```

```python
#in regression scaling is not needed
```

```python
df.describe()
```

```python
#splitting the dataset into training and test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.20, random_state=0)
```

```python
#fitting the simple linear regression model the training dataset

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#prediction of test and training set result

y_pred=regressor.predict(x_test)
x_pred=regressor.predict(x_train)
```

```python
#evaluating performance
```

```python
print(regressor.score(x_test, y_test))
```

```python
print(regressor.score(x_train, y_train))
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
from sklearn import metrics
print('mean absolute error :',metrics.mean_absolute_error(y_test, y_pred))
print('mean absolute error :',metrics.mean_squared_error(y_test, y_pred))
print('mean absolute error :',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
